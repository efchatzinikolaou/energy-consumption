```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example aggregated dataframe
# df = pd.read_csv('your_data.csv')

# Summing up annual consumption per city, year, and type
data_agg = df.groupby(['city', 'year', 'type'])['annual_consume'].sum().reset_index()

# Identifying the top 10 and bottom 10 cities by total consumption per year per type
top_10_cities_per_year_type = data_agg.groupby(['year', 'type']).apply(lambda x: x.nlargest(10, 'annual_consume')).reset_index(drop=True)
bottom_10_cities_per_year_type = data_agg.groupby(['year', 'type']).apply(lambda x: x.nsmallest(10, 'annual_consume')).reset_index(drop=True)

# Set the aesthetic style of the plots
sns.set_theme(style="whitegrid")

# Function to plot the data
def plot_city_consumption(data, title_prefix):
    for year in data['year'].unique():
        for consumption_type in data['type'].unique():
            subset = data[(data['year'] == year) & (data['type'] == consumption_type)]
            plt.figure(figsize=(10, 6))
            sns.barplot(x="city", y="annual_consume", data=subset, palette="deep", ci=None)
            plt.title(f"{title_prefix} - Year: {year} | Type: {consumption_type}")
            plt.xlabel("City")
            plt.ylabel("Total Consumption")
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.show()

# Plotting top 10 cities per year per type
plot_city_consumption(top_10_cities_per_year_type, "Top 10 Cities by Consumption")

# Plotting bottom 10 cities per year per type
plot_city_consumption(bottom_10_cities_per_year_type, "Bottom 10 Cities by Consumption")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample DataFrame
# Assuming the 'df' DataFrame is already loaded and contains the columns as specified.

# Step 1: Aggregate the data by city, year, and type (electricity or gas)
df_grouped = df.groupby(['city', 'year', 'type'], as_index=False)['annual_consume'].sum()

# Step 2: Separate the electricity and gas data
df_electricity = df_grouped[df_grouped['type'] == 'gas']
df_gas = df_grouped[df_grouped['type'] == 'electricity']
```

```python
# Step 2: Calculate total consumption per city (across both electricity and gas)
df_total_city_consumption = df_grouped.groupby('city')['annual_consume'].sum().reset_index()

# Step 3: Sort the cities by total consumption and get the top 10 cities
top_10_cities = df_total_city_consumption.sort_values(by='annual_consume', ascending=False).head(10)['city']

# Step 4: Filter the original grouped data to only include the top 10 cities
df_grouped_filtered = df_grouped[df_grouped['city'].isin(top_10_cities)]

# Step 5: Separate the electricity and gas data
df_electricity = df_grouped_filtered[df_grouped_filtered['type'] == 'electricity']
df_gas = df_grouped_filtered[df_grouped_filtered['type'] == 'gas']

# Step 6: Find the highest and lowest consumption per type (electricity and gas)
highest_electricity = df_electricity.loc[df_electricity['annual_consume'].idxmax()]
lowest_electricity = df_electricity.loc[df_electricity['annual_consume'].idxmin()]

highest_gas = df_gas.loc[df_gas['annual_consume'].idxmax()]
lowest_gas = df_gas.loc[df_gas['annual_consume'].idxmin()]

# Step 7: Plotting the data
# Create a figure with subplots for electricity and gas consumption
fig, axes = plt.subplots(2, 1, figsize=(12, 10))

# Plot for Electricity consumption
sns.barplot(data=df_electricity, x='city', y='annual_consume', hue='year', ax=axes[0])
axes[0].set_title('Electricity Consumption per City per Year (Top 10 Cities)')
axes[0].set_ylabel('Annual Consumption (Kwh)')
axes[0].tick_params(axis='x', rotation=45)

# Plot for Gas consumption
sns.barplot(data=df_gas, x='city', y='annual_consume', hue='year', ax=axes[1])
axes[1].set_title('Gas Consumption per City per Year (Top 10 Cities)')
axes[1].set_ylabel('Annual Consumption (m3)')
axes[1].tick_params(axis='x', rotation=45)

# Highlight highest and lowest consumption for both electricity and gas
for ax, highest, lowest, label in zip([axes[0], axes[1]], 
                                      [highest_electricity, highest_gas], 
                                      [lowest_electricity, lowest_gas], 
                                      ['Electricity', 'Gas']):
    ax.scatter(highest['city'], highest['annual_consume'], color='green', s=100, label=f'Highest {label}')
    ax.scatter(lowest['city'], lowest['annual_consume'], color='red', s=100, label=f'Lowest {label}')
    ax.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
